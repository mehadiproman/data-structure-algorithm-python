```python
#Implementation of binary search
## Function Definition
def binarySearch(arr, i, j, x):
    while i <= j:
        mid = i + (j-i)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            ##Recursion --> calling the same version again with a different set of parameters
            return binarySearch(arr, mid+1, j, x)
        else:
            return binarySearch(arr, i, mid-1, x)
        ## Searching element is not present in the array
        return -1
        


## Driver code
arr = [2, 5, 10, 14, 18, 22, 27, 35, 40, 59]
x = 14
i = 0
j = len(arr) -1
## Function call
result = binarySearch(arr, i, j, x)
print("Searching element present at the index: ", result) 
```
